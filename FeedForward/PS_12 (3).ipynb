## a. Import necessary packages

```python
from sklearn.preprocessing import LabelBinarizer
from sklearn.metrics import classification_report

from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import SGD
from tensorflow.keras import backend as K

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

## b. Load training and testing data

```python
train_data = pd.read_csv("C:/Users/Pranav/OneDrive/Desktop/Final_Practical/LP-IV-datasets/LP-IV-datasets/LP-IV-datasets/CIFR(Ass2&3)/train_data.csv",header=None)
test_data = pd.read_csv("C:/Users/Pranav/OneDrive/Desktop/Final_Practical/LP-IV-datasets/LP-IV-datasets/LP-IV-datasets/CIFR(Ass2&3)/test_data.csv",header=None)
```

```python
train_data.drop(0,inplace=True)
```

```python
x_train = train_data.drop(3072, axis=1)
```

```python
y_train = train_data[3072]
y_train.head()
```

```python
test_data.head()
```

```python
test_data.drop(0, inplace=True)
test_data
```

```python
x_test = test_data.drop(3072, axis=1)
x_test
```

```python
y_test = test_data[3072]
```

```python
y_test.head()
```

```python
x_train=np.array(x_train)
x_test=np.array(x_test)
y_train=np.array(y_train)
y_test=np.array(y_test)
```

```python
x_train = x_train.astype("float32") / 255.0
x_test = x_test.astype("float32") / 255.0
# y_train = y_train.astype("int32") / 255.0
# y_test = y_test.astype("int32") / 255.0
x_train.shape
```

```python
lb = LabelBinarizer()

y_train = lb.fit_transform(y_train.astype("int32"))
y_test = lb.transform(y_test.astype("int32"))
```

## Model building

```python
model = Sequential()
model.add(Dense(64, input_shape=(3072,), activation='sigmoid'))
model.add(Dense(32, activation='sigmoid'))
model.add(Dense(10, activation='softmax'))
```

```python
from keras.optimizers import Adam
sgd=SGD(0.01)
model.compile(loss='categorical_crossentropy',optimizer=sgd, metrics=['accuracy'])
h=model.fit(x_train, y_train, validation_data=(x_test, y_test),epochs=10)
```

```python
predictions = model.predict(x_test)
print(classification_report(y_test.argmax(axis=1),predictions.argmax(axis=1),target_names=[str(x) for x in lb.classes_]))
```

```python
plt.style.use("ggplot")
plt.figure()

plt.plot(h.history["loss"],label="loss")
# np.arange(0,epochs),
plt.plot(h.history["val_loss"],label="val_loss")
# np.arange(0,epochs),
plt.legend()
plt.plot()
```
